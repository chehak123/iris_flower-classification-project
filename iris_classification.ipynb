```python
iris=load_iris()
```

```python
x_train,x_test,y_train,y_test=train_test_split(iris['data'],iris['target'],random_state=0)
# x train has training data for iris.data and y train has for iris.target
# similarly x test has testing data for iris.data and y test has testing data for iris.target
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)                    # fitting the training set
# prediction=knn.predict(x_test)
```

```python
print(knn.score(x_test,y_test))
prediction=knn.predict([[3, 0.9, 0.3, 9]])
if prediction[0]==0:
    print("setosa")
elif prediction[0]==1:
    print("versicolor")
elif prediction[0]==2:
    print("virginica")
```
